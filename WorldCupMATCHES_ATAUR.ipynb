```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid')

#Pairplot
plt.figure(figsize=(10, 6))
sns.pairplot(dt)
plt.title('Pairplot of World Cup Matches')
plt.show()

```

```python
# Boxplot for Goals Scored by Home Team
plt.figure(figsize=(10, 5))
sns.boxplot(x='Home Team Goals', data=dt)
plt.title('Boxplot of Home Team Goals')
plt.xlabel('Home Team Goals')
plt.show()
```

```python
# Barplot
sns.barplot(x='Year', y='Attendance', data=dt)
plt.show()
```

```python
plt.hist(dt['Home Team Goals'])#histogram
```

```python
sns.histplot(x='Home Team Goals',data=dt)#histogram
```

```python
# Histogram of Home Team Goals
plt.figure(figsize=(10, 5))
sns.histplot(dt['Home Team Goals'], bins=10, kde=True)
plt.title('Histogram of Home Team Goals')
plt.xlabel('Home Team Goals')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=dt)#scatterplot is used to show ratio between 2 variables
plt.show()
```

```python
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=dt,hue='Half-time Home Goals')#scatterplot is used to show ratio between 2 variables
plt.show()
```

```python
dt['total_goals'] = dt['Home Team Goals'] + dt['Away Team Goals']
dt.head()
#we added a new column in data frame :)
```

```python
# Distribution Plot for Away Team Goals
plt.figure(figsize=(10, 5))
sns.displot(dt['Away Team Goals'])
plt.title('Distribution Plot of Away Team Goals')
plt.xlabel('Away Team Goals')
plt.ylabel('Density')
plt.show()
```

```python
#correlation : two categorical or numerical values are related or not
corr=dt[['Away Team Goals','Home Team Goals']].corr()
print(corr)
sns.heatmap(corr)
```
